# Ejercicio 7: 
Reduce las dimensiones con PCA, si consideras necesario.

```python
df.dropna(inplace = True)
df['X4']= df['X4'].replace(0, 2)
df['X3']= df['X3'].replace([0.0, 5.0, 6.0], 2)
```

```python
df_scale = pd.DataFrame(preprocessing.scale(df), columns = df.columns)
df_scale.head()
```

```python
pcs = PCA()
ACP = pcs.fit_transform(df_scale)
pcs_labels = [f'PC{i+1}' for i in range(len(df_scale.columns))]
ACP_df = pd.DataFrame(ACP,columns=pcs_labels)
ACP_df.head()
```

```python
var_pcs = pd.DataFrame(data = {'var':pcs.explained_variance_ratio_.round(2), 'varcum':pcs.explained_variance_ratio_.cumsum().round(2)})
var_pcs.index = pcs_labels
var_pcs.head(23)
```

```python
# Indica la importancia de las variables en cada componente
ACP_cl = pd.DataFrame(
    pcs.components_.round(2),
    columns= ACP_df.columns,
    index = df_scale.columns)
ACP_cl.iloc[:,:8].abs().idxmax()
```

```python
PC_components = np.arange(pcs.n_components_) + 1

_ = sns.set(style = 'whitegrid', 
            font_scale = 1.2
            )

fig, ax = plt.subplots(figsize=(10, 7))

_ = sns.barplot(x = PC_components, 
                y = pcs.explained_variance_ratio_, 
                color = 'b'
                )

_ = sns.lineplot(x = PC_components-1, 
                 y = np.cumsum(pcs.explained_variance_ratio_), 
                 color = 'black', 
                 linestyle = '-', 
                 linewidth = 2, 
                 marker = 'o', 
                 markersize = 8
                 )

plt.title('Scree Plot')
plt.xlabel('N-th Principal Component')
plt.ylabel('Variance Explained')
plt.ylim(0, 1)
plt.show()
```
